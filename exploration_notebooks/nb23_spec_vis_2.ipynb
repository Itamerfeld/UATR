```python
import os
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from scipy import signal


# os.chdir(os.path.expanduser("~/Projects/UATR"))

# from dev.utils import Audio

# %load_ext autoreload
# %autoreload 2
        
```

```python
data_file = "../data/scooter_example_1.wav"
data, fs = librosa.load(data_file, sr=None)
```

```python
# spec parameters
n_fft = 16384  
overlap = 0.5
hop_length = int(n_fft * (1 - overlap))
color_map = "magma"
crop_freq = 4000  # Hz

f, t, Sxx = signal.spectrogram(data, fs, window='hann', nperseg=n_fft, noverlap=hop_length, scaling='spectrum')
```

```python
plt.figure(figsize=(15, 8))
plt.pcolormesh(t, f, 10 * np.log10(Sxx + 1e-10), cmap=color_map)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('Spectrogram')
plt.colorbar(label='Power Spectral Density [dB]')
plt.ylim(0, crop_freq)  # Limit frequency axis to 2 kHz
plt.show()
```
